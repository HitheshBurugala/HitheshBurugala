```python
#Renaming the column for more readibilty.
# Here i would like to change the colums of my data frame to give more readibility.

# From '' sales channel  '' -----> to ----->'' mode of purchase ''
# From '' ship date '' ----> to ---->'' shipping date ''.
# From '' Units sold '' -----> to -----> '' Total Units sold ''

# data.rename(columns={'present column name': 'to be changed to column name'})

data.rename(columns={'Sales Channel': 'Mode Of Purchase'}, inplace=True)
data.rename(columns={'Ship Date': 'shipping date'}, inplace=True)
data.rename(columns={'Units Sold': 'Total Units sold'}, inplace=True)
data.head(2)
```
